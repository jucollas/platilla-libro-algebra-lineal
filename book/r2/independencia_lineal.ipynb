##  1. Subespacios en $\mathbb{R}^2$

###  Definici贸n

Un **subespacio vectorial** $S$ de $\mathbb{R}^2$ es un conjunto de vectores que cumple tres condiciones fundamentales:

1. Contiene al **vector nulo** $\mathbf{0} = (0, 0)$  
2. Es **cerrado bajo la suma**:  
   Si $\mathbf{u}, \mathbf{v} \in S$, entonces $\mathbf{u} + \mathbf{v} \in S$.  
3. Es **cerrado bajo la multiplicaci贸n por escalares**:  
   Si $\mathbf{v} \in S$ y $\alpha \in \mathbb{R}$, entonces $\alpha\mathbf{v} \in S$.

---

###  Ejemplos

- **Ejemplo 1:** El conjunto $S = \{(x, 0) \mid x \in \mathbb{R}\}$  
  representa el eje $x$.  
  Es un subespacio porque cumple las tres propiedades.

- **Ejemplo 2:** El conjunto $T = \{(x, 1) \mid x \in \mathbb{R}\}$  
  **no es un subespacio**, porque no contiene al vector nulo (ning煤n vector de la forma $(x, 1)$ es $(0,0)$).

---

###  Visualizaci贸n en c贸digo


```python
import numpy as np
import matplotlib.pyplot as plt

# Eje x como subespacio
x = np.linspace(-5, 5, 100)
y = np.zeros_like(x)

plt.figure(figsize=(6,6))
plt.plot(x, y, color='blue', label='Subespacio S (Eje x)')
plt.scatter(0, 0, color='red', label='Vector nulo (0,0)')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.legend()
plt.title("Ejemplo de subespacio en R虏: el eje x")
plt.grid(True)
plt.show()
```

##  2. Combinaciones lineales

###  Definici贸n

Una **combinaci贸n lineal** de vectores $\mathbf{v_1}, \mathbf{v_2}$ es cualquier vector que puede expresarse como:

$$
\mathbf{u} = \alpha_1 \mathbf{v_1} + \alpha_2 \mathbf{v_2}
$$

donde $\alpha_1, \alpha_2 \in \mathbb{R}$ son **escalares**.

> El conjunto de **todas las combinaciones lineales** de $\mathbf{v_1}$ y $\mathbf{v_2}$ forma un **subespacio generado** por ellos.

---

###  Interpretaci贸n geom茅trica

* Si $\mathbf{v_1}$ y $\mathbf{v_2}$ **no son proporcionales**, generan todo el plano $\mathbb{R}^2$.
* Si **s铆 son proporcionales**, las combinaciones lineales solo generan una **recta**.

---

###  Ejemplo con c贸digo




```python
v1 = np.array([1, 2])
v2 = np.array([2, 1])

# Generar combinaciones lineales
a = np.linspace(-2, 2, 10)
b = np.linspace(-2, 2, 10)
combinaciones = np.array([a_i * v1 + b_i * v2 for a_i in a for b_i in b])

# Visualizaci贸n
plt.figure(figsize=(6,6))
plt.scatter(combinaciones[:,0], combinaciones[:,1], color='blue', s=10, label='Combinaciones lineales')
plt.quiver(0,0,*v1,angles='xy',scale_units='xy',scale=1,color='red',label='v1')
plt.quiver(0,0,*v2,angles='xy',scale_units='xy',scale=1,color='green',label='v2')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.title("Subespacio generado por v1 y v2")
plt.legend()
plt.grid(True)
plt.show()
```


>  Si los puntos llenan todo el plano, entonces ${v_1, v_2}$ genera $\mathbb{R}^2$.

##  3. Independencia Lineal

###  Definici贸n

Un conjunto de vectores ${\mathbf{v_1}, \mathbf{v_2}}$ en $\mathbb{R}^2$ es **linealmente independiente** si la 煤nica soluci贸n de:

$$
\alpha_1\mathbf{v_1} + \alpha_2\mathbf{v_2} = \mathbf{0}
$$

es $\alpha_1 = \alpha_2 = 0$.

Si existen escalares no nulos que satisfacen la ecuaci贸n, los vectores son **dependientes lineales**.

---

###  Interpretaci贸n geom茅trica

* Dos vectores son **independientes** si **no son colineales** (no apuntan en la misma direcci贸n).
* Si uno es m煤ltiplo del otro, son **dependientes**, y el subespacio que generan es una **recta**, no todo $\mathbb{R}^2$.

---

###  Ejemplo en c贸digo


```python
import numpy as np

def son_independientes(v1, v2):
    """Determina si dos vectores en R2 son linealmente independientes."""
    matriz = np.column_stack((v1, v2))
    det = np.linalg.det(matriz)
    return not np.isclose(det, 0), det

v1 = np.array([1, 2])
v2 = np.array([2, 1])

indep, det = son_independientes(v1, v2)
print(f"驴Son independientes? {indep} (Determinante = {det:.2f})")
```



>  En $\mathbb{R}^2$, dos vectores son independientes **si su determinante es distinto de cero**.

---

###  Visualizaci贸n comparativa

```python
import matplotlib.pyplot as plt

v1 = np.array([1, 2])
v2 = np.array([2, 1])
v3 = np.array([2, 4])  # Dependiente de v1

plt.figure(figsize=(6,6))
plt.quiver(0,0,*v1,angles='xy',scale_units='xy',scale=1,color='red',label='v1')
plt.quiver(0,0,*v2,angles='xy',scale_units='xy',scale=1,color='green',label='v2 (independiente)')
plt.quiver(0,0,*v3,angles='xy',scale_units='xy',scale=1,color='blue',label='v3 (dependiente)')
plt.axhline(0,color='gray')
plt.axvline(0,color='gray')
plt.legend()
plt.title("Independencia vs Dependencia lineal en R虏")
plt.grid(True)
plt.show()
```

##  4. Relaci贸n entre los conceptos

| Concepto                 | Definici贸n                                                                        | Relaci贸n con los dem谩s                                                                       |
| :----------------------- | :-------------------------------------------------------------------------------- | :------------------------------------------------------------------------------------------- |
| **Subespacio**           | Conjunto cerrado bajo suma y multiplicaci贸n escalar.                              | El conjunto de todas las combinaciones lineales de un grupo de vectores forma un subespacio. |
| **Combinaci贸n lineal**   | Suma ponderada de vectores.                                                       | Genera nuevos vectores dentro de un subespacio.                                              |
| **Independencia lineal** | Ning煤n vector del conjunto puede expresarse como combinaci贸n lineal de los otros. | Determina si el subespacio tiene la **m谩xima dimensi贸n posible**.                            |

>  En $\mathbb{R}^2$:
>
> * Un solo vector no nulo genera una **recta** (subespacio 1D).
> * Dos vectores independientes generan **todo el plano** (subespacio 2D).

---

###  Ejemplo unificado




```python
v1 = np.array([1, 2])
v2 = np.array([2, 1])

# Verificar independencia
indep, det = son_independientes(v1, v2)

# Subespacio generado
a = np.linspace(-2, 2, 20)
b = np.linspace(-2, 2, 20)
combinaciones = np.array([a_i * v1 + b_i * v2 for a_i in a for b_i in b])

plt.figure(figsize=(6,6))
plt.scatter(combinaciones[:,0], combinaciones[:,1], color='lightblue', s=8, label='Subespacio generado')
plt.quiver(0,0,*v1,angles='xy',scale_units='xy',scale=1,color='red',label='v1')
plt.quiver(0,0,*v2,angles='xy',scale_units='xy',scale=1,color='green',label='v2')
plt.axhline(0,color='gray')
plt.axvline(0,color='gray')
plt.legend()
plt.title("Subespacio generado por v1 y v2" + ("\n R虏 completo" if indep else "\n Recta (dependientes)"))
plt.grid(True)
plt.show()
```
