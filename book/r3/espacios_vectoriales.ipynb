## üîπ 1. Representaci√≥n de vectores en $\mathbb{R}^3$

Un **vector en $\mathbb{R}^3$** se representa como un tr√≠o ordenado de n√∫meros reales:

$$
\mathbf{v} = (v_1, v_2, v_3), \quad v_i \in \mathbb{R}
$$

Cada componente indica cu√°nto se avanza en las tres direcciones ortogonales del espacio:

- $x$ ‚Üí direcci√≥n horizontal  
- $y$ ‚Üí direcci√≥n lateral  
- $z$ ‚Üí direcci√≥n vertical (profundidad)

Geom√©tricamente, un vector puede verse como una **flecha que parte del origen $(0,0,0)$** y termina en el punto $(v_1,v_2,v_3)$.

---

### üíª Ejemplo en c√≥digo: graficar un vector en 3D


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir el vector
v = np.array([2, 3, 1])

# Crear figura 3D
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar el vector desde el origen
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='red', arrow_length_ratio=0.1, linewidth=2)

# Configurar ejes
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)
ax.set_zlim(0, 4)
ax.set_xlabel('Eje X')
ax.set_ylabel('Eje Y')
ax.set_zlabel('Eje Z')
ax.set_title('Vector en $\mathbb{R}^3$')
plt.show()
```

## üîπ 2. Operaciones b√°sicas entre vectores

### ‚ûï 2.1 Suma de vectores

Dado $\mathbf{u} = (u_1, u_2, u_3)$ y $\mathbf{v} = (v_1, v_2, v_3)$:

$$
\mathbf{u} + \mathbf{v} = (u_1 + v_1,  u_2 + v_2,  u_3 + v_3)
$$

Esta operaci√≥n representa la **composici√≥n de desplazamientos** en el espacio.

---

### ‚úñÔ∏è 2.2 Multiplicaci√≥n por un escalar

Si $\alpha \in \mathbb{R}$, entonces:

$$
\alpha \mathbf{v} = (\alpha v_1, ; \alpha v_2, ; \alpha v_3)
$$

Multiplicar por un escalar **alarga, reduce o invierte** la direcci√≥n del vector.

---

### üíª Ejemplo en c√≥digo


```python
u = np.array([1, 2, 3])
v = np.array([3, -1, 2])

# Suma
suma = u + v

# Multiplicaci√≥n por escalar
alpha = 2
escalar_v = alpha * v

print("u + v =", suma)
print(f"{alpha} * v =", escalar_v)
```

---
### üí° Visualizaci√≥n de suma vectorial

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Dibujar vectores
ax.quiver(0, 0, 0, *u, color='blue', label='u', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, *v, color='green', label='v', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, *suma, color='red', label='u + v', arrow_length_ratio=0.1)

ax.set_xlim(-2, 5)
ax.set_ylim(-2, 5)
ax.set_zlim(-2, 5)
ax.legend()
ax.set_title("Suma de vectores en $\mathbb{R}^3$")
plt.show()
```

## üîπ 3. Norma de un vector

La **norma euclidiana** (longitud) de un vector en $\mathbb{R}^3$ es:

$$
|\mathbf{v}| = \sqrt{v_1^2 + v_2^2 + v_3^2}
$$

---

### üíª Ejemplo

```python
v = np.array([2, 3, 6])
norma_v = np.linalg.norm(v)
print(f"Norma de v: {norma_v:.2f}")
```

---

## üîπ 4. Producto punto (o escalar)

El **producto punto** entre dos vectores $\mathbf{u}$ y $\mathbf{v}$ se define como:

$$
\mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2 + u_3v_3
$$

y se relaciona con el **√°ngulo** $\theta$ entre ellos mediante:

$$
\mathbf{u} \cdot \mathbf{v} = |\mathbf{u}||\mathbf{v}|\cos(\theta)
$$

> üí° El producto punto mide **cu√°nto apuntan dos vectores en la misma direcci√≥n**.

---

### üíª Ejemplo en c√≥digo

```python
u = np.array([2, 0, 1])
v = np.array([1, 3, 2])

producto_punto = np.dot(u, v)
angulo = np.degrees(np.arccos(producto_punto / (np.linalg.norm(u) * np.linalg.norm(v))))

print("Producto punto:", producto_punto)
print(f"√Ångulo entre u y v: {angulo:.2f}¬∞")
```

---

## üîπ 5. Producto cruz (o vectorial)

El **producto cruz** entre dos vectores $\mathbf{u}$ y $\mathbf{v}$ produce un **nuevo vector perpendicular a ambos**:

$$
\mathbf{u} \times \mathbf{v} =
\begin{pmatrix}
u_2v_3 - u_3v_2, ;
u_3v_1 - u_1v_3, ;
u_1v_2 - u_2v_1
\end{pmatrix}
$$

Geom√©tricamente:

* $\mathbf{u} \times \mathbf{v}$ es **perpendicular** a ambos.
* Su magnitud representa el **√°rea del paralelogramo** formado por $\mathbf{u}$ y $\mathbf{v}$.

---

### üíª Ejemplo y visualizaci√≥n


```python
u = np.array([1, 0, 0])
v = np.array([0, 1, 0])
cruz = np.cross(u, v)

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

ax.quiver(0, 0, 0, *u, color='blue', label='u')
ax.quiver(0, 0, 0, *v, color='green', label='v')
ax.quiver(0, 0, 0, *cruz, color='red', label='u √ó v')

ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.legend()
ax.set_title("Producto cruz en $\mathbb{R}^3$")
plt.show()

print("u √ó v =", cruz)
```

## üîπ 6. Axiomas de un espacio vectorial

$\mathbb{R}^3$ es un **espacio vectorial real** porque cumple los siguientes **axiomas fundamentales**:

| Axioma | Propiedad                                                                                         |
| :----- | :------------------------------------------------------------------------------------------------ |
| 1      | $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$ (conmutatividad)                              |
| 2      | $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$ (asociatividad) |
| 3      | Existe $\mathbf{0}$ tal que $\mathbf{v} + \mathbf{0} = \mathbf{v}$                                |
| 4      | Para cada $\mathbf{v}$ existe $-\mathbf{v}$ tal que $\mathbf{v} + (-\mathbf{v}) = \mathbf{0}$     |
| 5      | $1\mathbf{v} = \mathbf{v}$                                                                        |
| 6      | $\alpha(\beta \mathbf{v}) = (\alpha\beta)\mathbf{v}$                                              |
| 7      | $\alpha(\mathbf{u} + \mathbf{v}) = \alpha\mathbf{u} + \alpha\mathbf{v}$                           |
| 8      | $(\alpha + \beta)\mathbf{v} = \alpha\mathbf{v} + \beta\mathbf{v}$                                 |

> ‚úÖ Estas propiedades garantizan que podemos operar libremente con vectores y escalares en $\mathbb{R}^3$.

## üíª 7. Implementaciones reutilizables


```python
def vector_norm(v):
    """Calcula la norma euclidiana de un vector."""
    return np.linalg.norm(v)

def dot_product(u, v):
    """Calcula el producto punto entre dos vectores."""
    return np.dot(u, v)

def cross_product(u, v):
    """Calcula el producto cruz entre dos vectores."""
    return np.cross(u, v)

def angle_between(u, v):
    """Calcula el √°ngulo entre dos vectores en grados."""
    cos_theta = np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
    return np.degrees(np.arccos(np.clip(cos_theta, -1, 1)))

# Ejemplo
u = np.array([2, 1, 0])
v = np.array([0, 1, 1])

print("Norma de u:", vector_norm(u))
print("Producto punto:", dot_product(u, v))
print("Producto cruz:", cross_product(u, v))
print("√Ångulo entre u y v:", angle_between(u, v))
```
