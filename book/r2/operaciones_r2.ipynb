En esta clase exploraremos las operaciones fundamentales que se pueden realizar con **vectores en el plano**.

## ðŸ”¹ 1. Vectores en $\mathbb{R}^2$

### ðŸ§© DefiniciÃ³n

Un **vector** en $\mathbb{R}^2$ es un par ordenado de nÃºmeros reales:

$$
\mathbf{v} = (v_1, v_2), \quad v_1, v_2 \in \mathbb{R}.
$$

TambiÃ©n puede escribirse en **forma columna**:

$$
\mathbf{v} =
\begin{pmatrix}
v_1 \\[4pt]
v_2
\end{pmatrix}
$$

Cada componente indica cuÃ¡nto se avanza en los ejes $x$ y $y$ respectivamente.

---

### ðŸ“ˆ InterpretaciÃ³n geomÃ©trica

Un vector se representa como una **flecha** que parte del **origen** $(0,0)$ y llega al punto $(v_1,v_2)$.

- Su **longitud** indica la magnitud del desplazamiento.  
- Su **direcciÃ³n** indica hacia dÃ³nde apunta el vector.

> ðŸ’¡ **Recuerda:** un punto representa una **posiciÃ³n**, mientras que un vector representa un **desplazamiento o cambio**.

---

## âž• 2. Suma de vectores

### ðŸ“˜ DefiniciÃ³n

Sean dos vectores:
$$
\mathbf{u} = (u_1, u_2), \quad \mathbf{v} = (v_1, v_2),
$$
la **suma vectorial** se define como:
$$
\mathbf{u} + \mathbf{v} = (u_1 + v_1,\; u_2 + v_2).
$$

---

### ðŸ§­ InterpretaciÃ³n geomÃ©trica

Existen dos formas equivalentes de visualizar la suma de vectores:

- **Regla del paralelogramo:**  
  Coloca ambos vectores con el mismo origen y forma un paralelogramo; la diagonal representa la suma.

- **MÃ©todo cabeza-cola:**  
  Traslada el vector $\mathbf{v}$ de modo que su cola coincida con la cabeza de $\mathbf{u}$.  
  El vector resultante $\mathbf{u} + \mathbf{v}$ va desde la cola de $\mathbf{u}$ hasta la cabeza de $\mathbf{v}$.

---

### ðŸ§® Ejemplo

$$
\mathbf{u} = (2, 3), \quad \mathbf{v} = (-1, 5)
$$

$$
\mathbf{u} + \mathbf{v} = (2 + (-1),\; 3 + 5) = (1, 8)
$$

---

### ðŸ’» VisualizaciÃ³n (Python)


```python
import matplotlib.pyplot as plt

u = (2, 3)
v = (-1, 5)
suma = (u[0] + v[0], u[1] + v[1])

plt.figure(figsize=(5,5))
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.quiver(0, 0, *u, angles='xy', scale_units='xy', scale=1, color='blue', label='u')
plt.quiver(0, 0, *v, angles='xy', scale_units='xy', scale=1, color='green', label='v')
plt.quiver(0, 0, *suma, angles='xy', scale_units='xy', scale=1, color='red', label='u + v')

plt.xlim(-2, 5)
plt.ylim(-2, 9)
plt.grid(True)
plt.legend()
plt.title("Suma de vectores en RÂ²")
plt.show()
```

## âœ–ï¸ 3. MultiplicaciÃ³n por un escalar

### ðŸ“˜ DefiniciÃ³n

Si $\mathbf{v} = (v_1, v_2)$ y $\alpha \in \mathbb{R}$ es un **escalar**, entonces:

$$
\alpha \mathbf{v} = (\alpha v_1, ; \alpha v_2)
$$

---

### ðŸ“ˆ InterpretaciÃ³n geomÃ©trica

* Multiplicar un vector por un escalar **cambia su longitud**.
* Si $\alpha > 1$, el vector se **alarga**.
* Si $0 < \alpha < 1$, el vector se **encoge**.
* Si $\alpha < 0$, ademÃ¡s **invierte su direcciÃ³n**.

---

### ðŸ§® Ejemplo

$$
\mathbf{v} = (2, -1), \quad \alpha = 3
$$

$$
3\mathbf{v} = (6, -3)
$$

---

### ðŸ’» VisualizaciÃ³n (Python)



```python
v = (2, -1)
alpha = 3
scaled = (alpha * v[0], alpha * v[1])

plt.figure(figsize=(5,5))
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')

plt.quiver(0, 0, *v, angles='xy', scale_units='xy', scale=1, color='blue', label='v')
plt.quiver(0, 0, *scaled, angles='xy', scale_units='xy', scale=1, color='red', label='3v')

plt.xlim(-1, 7)
plt.ylim(-4, 2)
plt.grid(True)
plt.legend()
plt.title("MultiplicaciÃ³n por escalar")
plt.show()
```

## âš–ï¸ 4. Propiedades de las operaciones en $\mathbb{R}^2$

Las operaciones de **suma de vectores** y **multiplicaciÃ³n por escalares** cumplen las propiedades bÃ¡sicas de los espacios vectoriales:

| Propiedad                       | ExpresiÃ³n                                                                         | InterpretaciÃ³n                              |
| :------------------------------ | :-------------------------------------------------------------------------------- | :------------------------------------------ |
| **Conmutatividad**              | $\mathbf{u} + \mathbf{v} = \mathbf{v} + \mathbf{u}$                               | El orden de la suma no cambia el resultado. |
| **Asociatividad**               | $(\mathbf{u} + \mathbf{v}) + \mathbf{w} = \mathbf{u} + (\mathbf{v} + \mathbf{w})$ | Podemos agrupar libremente las sumas.       |
| **Elemento neutro**             | $\mathbf{v} + \mathbf{0} = \mathbf{v}$                                            | Sumar el vector nulo no cambia nada.        |
| **Inverso aditivo**             | $\mathbf{v} + (-\mathbf{v}) = \mathbf{0}$                                         | Todo vector tiene un opuesto.               |
| **Distributividad (escalares)** | $\alpha(\mathbf{u} + \mathbf{v}) = \alpha\mathbf{u} + \alpha\mathbf{v}$           | El escalar se reparte en la suma.           |
| **Asociatividad de escalares**  | $\alpha(\beta \mathbf{v}) = (\alpha\beta)\mathbf{v}$                              | Multiplicar escalares es compatible.        |
| **Identidad multiplicativa**    | $1\mathbf{v} = \mathbf{v}$                                                        | Multiplicar por 1 no altera el vector.      |

## ðŸ“ 5. Normas de un vector

Una **norma** mide la â€œlongitudâ€ o â€œmagnitudâ€ de un vector.

### ðŸ”¸ Norma Euclidiana ($L_2$)

$$
|\mathbf{v}|_2 = \sqrt{v_1^2 + v_2^2}
$$

**Ejemplo:**
$$
\mathbf{v} = (3, 4) \Rightarrow
|\mathbf{v}|_2 = \sqrt{3^2 + 4^2} = 5
$$

---

### ðŸ”¸ Otras normas comunes

* **Norma 1 ($L_1$):**
  $$
  |\mathbf{v}|_1 = |v_1| + |v_2|
  $$
  Mide la â€œdistancia en callesâ€ o â€œdistancia Manhattanâ€.

* **Norma infinito ($L_\infty$):**
  $$
  |\mathbf{v}|_\infty = \max(|v_1|, |v_2|)
  $$
  Mide el mayor desplazamiento en cualquiera de las direcciones.

> ðŸ’¡ En geometrÃ­a, la norma Euclidiana ($L_2$) es la mÃ¡s usada,
> pero las otras normas son muy Ãºtiles en **optimizaciÃ³n**, **aprendizaje automÃ¡tico** y **ciencias de datos**.

---

## âš–ï¸ 6. Propiedades de la Norma

La **norma euclidiana** en $\mathbb{R}^2$ cumple con propiedades fundamentales que la convierten en una medida coherente de longitud o magnitud.

Sea $\mathbf{u}, \mathbf{v} \in \mathbb{R}^2$ y $\alpha \in \mathbb{R}$.

---

### ðŸ”¹ 1. Positividad

$$
\|\mathbf{v}\|_2 \geq 0
\quad \text{y} \quad
\|\mathbf{v}\|_2 = 0 \text{ si y solo si } \mathbf{v} = \mathbf{0}.
$$

> La norma nunca es negativa.  
> Solo el **vector nulo** tiene longitud cero.

---

### ðŸ”¹ 2. Homogeneidad

$$
\|\alpha \mathbf{v}\|_2 = |\alpha| \, \|\mathbf{v}\|_2.
$$

> Multiplicar por un escalar **escala la longitud** del vector por el valor absoluto del escalar.  
> Si $\alpha < 0$, el vector cambia de direcciÃ³n pero mantiene la misma magnitud.

---

### ðŸ”¹ 3. Desigualdad Triangular

$$
\|\mathbf{u} + \mathbf{v}\|_2 \leq \|\mathbf{u}\|_2 + \|\mathbf{v}\|_2.
$$

> La longitud de la suma de dos vectores **no puede exceder** la suma de sus longitudes individuales.  
> GeomÃ©tricamente, esto significa que el lado de un triÃ¡ngulo nunca es mÃ¡s largo que la suma de los otros dos.

---

### ðŸ’¡ VisualizaciÃ³n (Python)



```python
import matplotlib.pyplot as plt
import numpy as np

u = np.array([2, 1])
v = np.array([1, 3])
sum_vec = u + v

plt.figure(figsize=(5,5))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)
plt.quiver(0,0,*u,angles='xy',scale_units='xy',scale=1,color='blue',label='u')
plt.quiver(0,0,*v,angles='xy',scale_units='xy',scale=1,color='green',label='v')
plt.quiver(0,0,*sum_vec,angles='xy',scale_units='xy',scale=1,color='red',label='u+v')

plt.legend()
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.title("Desigualdad triangular en RÂ²")
plt.grid(True)
plt.show()
```

## ðŸŽ¯ 7. ProyecciÃ³n de un Vector en $\mathbb{R}^2$

La **proyecciÃ³n** de un vector $\mathbf{u}$ sobre otro vector $\mathbf{v}$ (con $\mathbf{v} \neq \mathbf{0}$) representa **la parte de $\mathbf{u}$ que apunta en la direcciÃ³n de $\mathbf{v}$**.

GeomÃ©tricamente, es el **vector sombra** de $\mathbf{u}$ sobre $\mathbf{v}$.

---

### ðŸ“˜ 7.1 DefiniciÃ³n

$$
\text{proj}_{\mathbf{v}}(\mathbf{u}) =
\left( \frac{\mathbf{u} \cdot \mathbf{v}}{\mathbf{v} \cdot \mathbf{v}} \right) \mathbf{v}
$$

Donde:

* El **producto punto** se define como
  $$
  \mathbf{u} \cdot \mathbf{v} = u_1v_1 + u_2v_2
  $$
* Y
  $$
  \mathbf{v} \cdot \mathbf{v} = |\mathbf{v}|_2^2 = v_1^2 + v_2^2
  $$

---

### ðŸ§® 7.2 Ejemplo

Sean:
$$
\mathbf{u} = (2, 1), \quad \mathbf{v} = (3, 0)
$$

1. **Producto punto:**
   $$
   \mathbf{u} \cdot \mathbf{v} = (2)(3) + (1)(0) = 6
   $$
2. **Producto de $\mathbf{v}$ consigo misma:**
   $$
   \mathbf{v} \cdot \mathbf{v} = (3)(3) + (0)(0) = 9
   $$
3. **ProyecciÃ³n:**
   $$
   \text{proj}_{\mathbf{v}}(\mathbf{u})
   = \left( \frac{6}{9} \right)(3, 0)
   = \left( \frac{2}{3} \right)(3, 0)
   = (2, 0)
   $$

Por tanto, la proyecciÃ³n de $\mathbf{u}$ sobre $\mathbf{v}$ es el vector **$(2, 0)$**, que apunta en la misma direcciÃ³n que $\mathbf{v}$.

---

### ðŸ’» VisualizaciÃ³n (Python)



```python
import numpy as np
import matplotlib.pyplot as plt

u = np.array([2, 1])
v = np.array([3, 0])
proj = (np.dot(u, v) / np.dot(v, v)) * v

plt.figure(figsize=(6,5))
plt.axhline(0, color='gray', linewidth=1)
plt.axvline(0, color='gray', linewidth=1)

plt.quiver(0,0,*u,angles='xy',scale_units='xy',scale=1,color='blue',label='u')
plt.quiver(0,0,*v,angles='xy',scale_units='xy',scale=1,color='green',label='v')
plt.quiver(0,0,*proj,angles='xy',scale_units='xy',scale=1,color='red',label='proj_v(u)')

plt.xlim(-1,4)
plt.ylim(-1,3)
plt.legend()
plt.title("ProyecciÃ³n de u sobre v en RÂ²")
plt.grid(True)
plt.show()
```


> ðŸ’¬ Observa cÃ³mo la proyecciÃ³n **cae sobre la misma lÃ­nea** que $\mathbf{v}$ y **reduce la dimensiÃ³n** del problema a una sola direcciÃ³n.

## ðŸ§© Implementaciones Reutilizables en Python para $\mathbb{R}^2$

En esta secciÃ³n encontrarÃ¡s un conjunto de **funciones reutilizables** que implementan las operaciones fundamentales vistas en el capÃ­tulo:  
**suma de vectores, multiplicaciÃ³n por escalar, norma y proyecciÃ³n**, junto con **rutinas de visualizaciÃ³n grÃ¡fica**.

Estas funciones te permitirÃ¡n **experimentar, visualizar y comprobar** los conceptos teÃ³ricos directamente en cÃ³digo.

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------
# ðŸ§® Funciones MatemÃ¡ticas BÃ¡sicas
# ----------------------------

def vector_addition(u, v):
    """Suma de dos vectores en R2"""
    return np.array([u[0] + v[0], u[1] + v[1]])

def scalar_multiplication(alpha, v):
    """MultiplicaciÃ³n de un vector por un escalar"""
    return np.array([alpha * v[0], alpha * v[1]])

def vector_norm(v):
    """Calcula la norma euclidiana de un vector"""
    return np.sqrt(v[0]**2 + v[1]**2)

def vector_projection(u, v):
    """ProyecciÃ³n del vector u sobre v"""
    v_norm_squared = v[0]**2 + v[1]**2
    scalar_projection = np.dot(u, v) / v_norm_squared
    return scalar_projection * v
```

```python
# ----------------------------
# ðŸŽ¨ Funciones de VisualizaciÃ³n
# ----------------------------

def plot_vector_addition(u, v):
    """Representa grÃ¡ficamente la suma de vectores usando el mÃ©todo cabeza-cola"""
    result = vector_addition(u, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Dibujar los vectores
    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='v')
    plt.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1, color='red', label='u + v')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('Suma de vectores (MÃ©todo cabeza-cola)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


def plot_scalar_multiplication(alpha, v):
    """Representa grÃ¡ficamente la multiplicaciÃ³n por escalar"""
    scaled_v = scalar_multiplication(alpha, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v')
    plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'{alpha}v')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('MultiplicaciÃ³n por escalar')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()


def plot_vector_projection(u, v):
    """Representa grÃ¡ficamente la proyecciÃ³n de un vector sobre otro"""
    proj = vector_projection(u, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='v')
    plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='red', label='proj_v(u)')

    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('ProyecciÃ³n de un vector sobre otro')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

```python
# ----------------------------
# ðŸ§© Ejemplo de uso
# ----------------------------

if __name__ == "__main__":
    # Definir vectores
    u = np.array([3, 2])
    v = np.array([1, 4])

    print("Vector u:", u)
    print("Vector v:", v)

    # Suma de vectores
    u_plus_v = vector_addition(u, v)
    print("Suma de vectores u + v:", u_plus_v)

    # MultiplicaciÃ³n por escalar
    alpha = 2
    scaled_u = scalar_multiplication(alpha, u)
    print(f"Vector u multiplicado por {alpha}:", scaled_u)

    # Norma de un vector
    norm_u = vector_norm(u)
    print("Norma del vector u:", norm_u)

    # ProyecciÃ³n de u sobre v
    proj_u_on_v = vector_projection(u, v)
    print("ProyecciÃ³n de u sobre v:", proj_u_on_v)

    # Visualizaciones
    plot_vector_addition(u, v)
    plot_scalar_multiplication(alpha, u)
    plot_vector_projection(u, v)
```
