##  1. 驴Qu茅 es una recta en $\mathbb{R}^2$?

Una **recta** en el plano $\mathbb{R}^2$ es el conjunto de todos los puntos que se alinean en una misma direcci贸n.  
Puede describirse de dos formas principales:

- **Forma cartesiana:** mediante una ecuaci贸n lineal entre $x$ e $y$.  
- **Forma vectorial o param茅trica:** mediante un punto y un vector direcci贸n.

---

###  1.1 Forma cartesiana

La **ecuaci贸n cartesiana** general de una recta es:

$$
Ax + By + C = 0
$$

donde $A$, $B$, y $C$ son constantes reales.

- Si $B \neq 0$, se puede despejar $y$:
  $$
  y = -\frac{A}{B}x - \frac{C}{B}
  $$
  La pendiente de la recta es $m = -\frac{A}{B}$.

- Si $B = 0$, la recta es **vertical** y su ecuaci贸n es simplemente $x = -\frac{C}{A}$.

---

###  Ejemplo en c贸digo




```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes de la recta Ax + By + C = 0
A, B, C = 2, -3, 6

# Crear puntos para graficar
x = np.linspace(-10, 10, 100)
y = -(A/B)*x - C/B

plt.figure(figsize=(7,5))
plt.plot(x, y, label=f"{A}x + ({B})y + {C} = 0")
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.legend()
plt.title("Recta en forma cartesiana")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

##  2. Forma vectorial o param茅trica

La recta tambi茅n puede representarse **usando vectores**:

$$
\mathbf{r} = \mathbf{r_0} + t\mathbf{v}
$$

donde:

* $\mathbf{r_0} = (x_0, y_0)$ es un **punto** de la recta.
* $\mathbf{v} = (a, b)$ es un **vector direcci贸n** de la recta.
* $t \in \mathbb{R}$ es un **par谩metro** real.

Esta forma permite generar todos los puntos de la recta variando el valor de $t$.

---

###  Ejemplo en c贸digo



```python
# Punto y vector direcci贸n
r0 = np.array([1, 2])
v = np.array([2, 1])

# Generar puntos sobre la recta para distintos valores de t
t = np.linspace(-5, 5, 50)
x = r0[0] + v[0]*t
y = r0[1] + v[1]*t

plt.figure(figsize=(7,5))
plt.plot(x, y, color='blue', label='Recta param茅trica')
plt.scatter(*r0, color='red', label='Punto inicial r0')
plt.quiver(r0[0], r0[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector direcci贸n')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.legend()
plt.title("Recta en forma vectorial")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

##  3. Conversi贸n entre formas

### 3.1 De forma cartesiana a forma param茅trica

Dada una recta:

$$
Ax + By + C = 0
$$

* Escogemos un **punto particular** que cumpla la ecuaci贸n.
  Por ejemplo, si $B \neq 0$, podemos tomar $x_0 = 0$ y calcular $y_0 = -\frac{C}{B}$.
* El **vector direcci贸n** $\mathbf{v}$ es perpendicular al vector normal $(A, B)$, por lo que se puede tomar:
  $$
  \mathbf{v} = (B, -A)
  $$

Entonces la forma param茅trica queda:

$$
(x, y) = (x_0, y_0) + t(B, -A)
$$

---

###  Ejemplo




```python
A, B, C = 2, -3, 6

# Punto particular sobre la recta (x0, y0)
x0 = 0
y0 = -C/B

# Vector direcci贸n
v = np.array([B, -A])
r0 = np.array([x0, y0])

print("Punto sobre la recta:", r0)
print("Vector direcci贸n:", v)

# Graficar
t = np.linspace(-5, 5, 50)
x = r0[0] + v[0]*t
y = r0[1] + v[1]*t

plt.figure(figsize=(7,5))
plt.plot(x, y, color='blue', label='Forma param茅trica de la recta')
plt.scatter(*r0, color='red', label='Punto r0')
plt.quiver(r0[0], r0[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='Vector direcci贸n')
plt.axhline(0, color='gray')
plt.axvline(0, color='gray')
plt.legend()
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Conversi贸n de forma cartesiana a param茅trica")
plt.show()
```

### 3.2 De forma param茅trica a cartesiana

Si una recta est谩 dada por:

$$
(x, y) = (x_0, y_0) + t(a, b)
$$

entonces:

$$
\begin{cases}
x = x_0 + at \
y = y_0 + bt
\end{cases}
$$

De aqu铆 se despeja $t = \frac{x - x_0}{a}$ y se sustituye en la segunda ecuaci贸n:

$$
y - y_0 = \frac{b}{a}(x - x_0)
$$

Finalmente se obtiene:

$$
b(x - x_0) - a(y - y_0) = 0
$$

que es la **forma cartesiana** de la misma recta.

---

###  Ejemplo


```python
# Forma param茅trica
r0 = np.array([1, -2])
v = np.array([2, 3])

# Coeficientes cartesianos A, B, C
A = -v[1]
B = v[0]
C = -(A * r0[0] + B * r0[1])

print(f"Ecuaci贸n cartesiana: {A}x + {B}y + {C} = 0")
```

##  4. Distancia entre dos puntos

La **distancia euclidiana** entre dos puntos $P_1 = (x_1, y_1)$ y $P_2 = (x_2, y_2)$ se calcula como:

$$
d(P_1, P_2) = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}
$$

---

###  Ejemplo



```python
P1 = np.array([2, 3])
P2 = np.array([-1, 7])

distancia = np.linalg.norm(P2 - P1)
print(f"Distancia entre P1 y P2: {distancia:.2f}")
```

##  5. ngulo entre dos vectores

El **谩ngulo** $\theta$ entre dos vectores $\mathbf{u}$ y $\mathbf{v}$ se define mediante el producto punto:

$$
\mathbf{u} \cdot \mathbf{v} = |\mathbf{u}||\mathbf{v}|\cos(\theta)
$$

De donde:

$$
\theta = \arccos\left(\frac{\mathbf{u} \cdot \mathbf{v}}{|\mathbf{u}||\mathbf{v}|}\right)
$$

---

###  Ejemplo


```python
u = np.array([3, 4])
v = np.array([4, -2])

# Calcular el 谩ngulo en radianes y grados
dot_product = np.dot(u, v)
norm_u = np.linalg.norm(u)
norm_v = np.linalg.norm(v)
theta_rad = np.arccos(dot_product / (norm_u * norm_v))
theta_deg = np.degrees(theta_rad)

print(f"ngulo entre u y v: {theta_deg:.2f}掳")
```
