# üìò Operaciones en $\mathbb{R}^2$

En esta clase repasaremos los conceptos fundamentales de los **vectores en $\mathbb{R}^2$**, la **suma de vectores**, la **multiplicaci√≥n por escalar**, las **normas** y la **proyecci√≥n de un vector sobre otro**.

## üîπ Vectores en $\mathbb{R}^2$

### Definici√≥n
Un **vector** en $\mathbb{R}^2$ es un par ordenado de n√∫meros reales:
$$
\mathbf{v} = (v_1, v_2), \quad v_1,v_2 \in \mathbb{R}.
$$

Tambi√©n se puede escribir en forma de columna:
$$
\mathbf{v}=\begin{pmatrix}v_1\\v_2\end{pmatrix}.
$$

### Interpretaci√≥n Geom√©trica
- Se representa como un **segmento dirigido** en el plano cartesiano desde el origen $(0,0)$ hasta $(v_1,v_2)$.  
- Su **longitud** y **direcci√≥n** describen el vector.

---
## ‚ûï Suma de Vectores

### Definici√≥n
Dadas dos vectores en el plano
$$
\mathbf{u}=(u_1,u_2), \quad \mathbf{v}=(v_1,v_2),
$$
su **suma** se define como:
$$
\mathbf{u}+\mathbf{v}=(u_1+v_1,\;u_2+v_2).
$$

### Interpretaci√≥n Geom√©trica
Para sumar $\mathbf{u}$ y $\mathbf{v}$ en el plano se puede usar:

- **Regla del paralelogramo**: colocar ambos vectores con el mismo punto de partida y formar un paralelogramo; la diagonal representa la suma.  
- **M√©todo cabeza-cola**: trasladar (sin rotar) el vector $\mathbf{v}$ de modo que su cola coincida con la cabeza de $\mathbf{u}$. El vector resultante va desde la cola de $\mathbf{u}$ hasta la cabeza de $\mathbf{v}$.

### Ejemplo
Si
$$
\mathbf{u}=(2,3),\quad \mathbf{v}=(-1,5),
$$
entonces:
$$
\mathbf{u}+\mathbf{v}=(2+(-1),\;3+5)=(1,8).
$$

---

## ‚úñÔ∏è Multiplicaci√≥n por un Escalar

### Definici√≥n
Dado un vector $\mathbf{v}=(v_1,v_2)$ y un n√∫mero real $\alpha \in \mathbb{R}$ (llamado **escalar**), la multiplicaci√≥n se define como:
$$
\alpha \mathbf{v}=(\alpha v_1,\;\alpha v_2).
$$

### Interpretaci√≥n Geom√©trica
- Multiplicar un vector por un escalar **estira** o **encoge** al vector.  
- Si $\alpha<0$, adem√°s de cambiar su longitud, **invierte su direcci√≥n**.  

### Ejemplo
Sea $\mathbf{v}=(2,-1)$ y $\alpha=3$. Entonces:
$$
3\mathbf{v}=3\cdot(2,-1)=(6,-3).
$$

---

## ‚öñÔ∏è Propiedades en $\mathbb{R}^2$

Las operaciones de suma de vectores y multiplicaci√≥n por escalares cumplen las siguientes propiedades:

1. **Conmutatividad de la suma**  
   $$
   \mathbf{u}+\mathbf{v}=\mathbf{v}+\mathbf{u}.
   $$
2. **Asociatividad de la suma**  
   $$
   (\mathbf{u}+\mathbf{v})+\mathbf{w}=\mathbf{u}+(\mathbf{v}+\mathbf{w}).
   $$
3. **Elemento neutro**  
   $$
   \mathbf{v}+\mathbf{0}=\mathbf{v}, \quad \mathbf{0}=(0,0).
   $$
4. **Inverso aditivo**  
   $$
   \mathbf{v}+(-\mathbf{v})=\mathbf{0}.
   $$
5. **Distributividad**  
   $$
   \alpha(\mathbf{u}+\mathbf{v})=\alpha\mathbf{u}+\alpha\mathbf{v}.
   $$
6. **Asociatividad de escalares**  
   $$
   \alpha(\beta\mathbf{v})=(\alpha\beta)\mathbf{v}.
   $$
7. **Identidad multiplicativa**  
   $$
   1\mathbf{v}=\mathbf{v}.
   $$

## üìè Normas de un Vector

La **norma** mide la ‚Äúlongitud‚Äù o ‚Äúmagnitud‚Äù de un vector.

### Norma Euclidiana ($L_2$)
Para $\mathbf{v}=(v_1,v_2)$:
$$
\|\mathbf{v}\|_2=\sqrt{v_1^2+v_2^2}.
$$

- Ejemplo:  
  Si $\mathbf{v}=(3,4)$,
  $$
  \|\mathbf{v}\|_2=\sqrt{3^2+4^2}=\sqrt{9+16}=\sqrt{25}=5.
  $$

### Otras Normas
- **Norma 1 ($L_1$)**:  
  $$
  \|\mathbf{v}\|_1=|v_1|+|v_2|.
  $$
- **Norma infinito ($L_\infty$)**:  
  $$
  \|\mathbf{v}\|_\infty=\max(|v_1|,|v_2|).
  $$

üìå La norma euclidiana ($L_2$) es la m√°s usada en geometr√≠a y an√°lisis vectorial, pero las otras normas son √∫tiles en optimizaci√≥n y ciencias de datos.
 

```python
import numpy as np
import matplotlib.pyplot as plt

valores_de_p = [1,2]
colores = ['red','blue']

x = np.linspace(-1,1,1000)

plt.figure(figsize=(8,6))
for j,p in enumerate(valores_de_p):
    y = (1 - np.abs(x)**p)**(1/p)
    plt.plot(x,y,color=colores[j],label=f"L_{p}")
    plt.plot(x,-y,color=colores[j])
plt.axis("equal")
plt.grid(True)
plt.legend()
plt.title("Circunferencias en diferentes normas")
plt.show()
```

### Propiedades de la Norma

La norma euclidiana en $\mathbb{R}^2$ cumple con las siguientes propiedades:

1. **Positividad**:
   $$
   \|\mathbf{v}\|_2\geq 0,\quad\text{y}\quad\|\mathbf{v}\|_2=0\text{ si y solo si }\mathbf{v}=\mathbf{0}.
   $$
2. **Homogeneidad**:
    $$
    \|\alpha\mathbf{v}\|_2=|\alpha|\|\mathbf{v}\|_2.
    $$
3. **Desigualdad Triangular**:
    $$
    \|\mathbf{u}+\mathbf{v}\|_2\leq\|\mathbf{u}\|_2+\|\mathbf{v}\|_2.
    $$


## Proyecci√≥n de un Vector en $\mathbb{R}^2$

La **proyecci√≥n** de un vector $\mathbf{u}$ sobre otro vector $\mathbf{v}$ (asumiendo $\mathbf{v}\neq\mathbf{0}$) es el vector en la direcci√≥n de $\mathbf{v}$ que representa la parte de $\mathbf{u}$ en esa direcci√≥n.

### 5.1 Definici√≥n
La proyecci√≥n de $\mathbf{u}$ sobre $\mathbf{v}$, denotada como $\text{proj}_{\mathbf{v}}(\mathbf{u})$, se define como:
$$
\text{proj}_{\mathbf{v}}(\mathbf{u})=\left(\frac{\mathbf{u}\cdot\mathbf{v}}{\mathbf{v}\cdot\mathbf{v}}\right)\mathbf{v}.
$$
- $\mathbf{u}\cdot\mathbf{v}$ es el **producto punto**:
  $$
  \mathbf{u}\cdot\mathbf{v}=u_1v_1+u_2v_2.
  $$
- $\mathbf{v}\cdot\mathbf{v}=\|\mathbf{v}\|_2^2=v_1^2+v_2^2$.

### 5.2 Ejemplo
Sean:
$$
\mathbf{u}=(2,1),\mathbf{v}=(3,0).
$$
1. Calcular el producto punto $\mathbf{u}\cdot\mathbf{v}$:
   $$
   \mathbf{u}\cdot\mathbf{v}=(2)(3)+(1)(0)=6.
   $$
2. Calcular $\mathbf{v}\cdot\mathbf{v}$:
   $$
   \mathbf{v}\cdot\mathbf{v}=(3)(3)+(0)(0)=9.
   $$
3. Hallar la proyecci√≥n:
   $$
   \text{proj}_{\mathbf{v}}(\mathbf{u})=\left(\frac{6}{9}\right)(3,0)=\left(\frac{2}{3}\right)(3,0)=(2,0).
   $$

Geom√©tricamente, $\mathbf{u}$ se proyecta sobre $\mathbf{v}$ y el vector resultante es **paralelo** a $\mathbf{v}$.

## 6. Resumen

1. **Vectores en $\mathbb{R}^2$:** Se representan como pares ordenados $(v_1,v_2)$.
2. **Suma de Vectores:** Suma componente a componente.
3. **Multiplicaci√≥n por Escalar:** Escala la magnitud del vector y puede invertir su direcci√≥n (si el escalar es negativo).
4. **Normas:** Principalmente la norma euclidiana $\sqrt{v_1^2+v_2^2}$ para medir la ‚Äúlongitud‚Äù del vector.
5. **Proyecci√≥n:** $\text{proj}_{\mathbf{v}}(\mathbf{u})$ calcula la componente de $\mathbf{u}$ en la direcci√≥n de $\mathbf{v}$.

Con estos conceptos, ya tienes las bases para entender y trabajar con operaciones vectoriales fundamentales en $\mathbb{R}^2$. ¬°Sigue practicando con ejemplos para reforzar tu comprensi√≥n!






```python
import numpy as np
import matplotlib.pyplot as plt

# Definimos las funciones

def vector_addition(u, v):
    """Suma de dos vectores en R2"""
    return np.array([u[0] + v[0], u[1] + v[1]])

def scalar_multiplication(alpha, v):
    """Multiplicaci√≥n de un vector por un escalar"""
    return np.array([alpha * v[0], alpha * v[1]])

def vector_norm(v):
    """Calcula la norma euclidiana de un vector"""
    return np.sqrt(v[0]**2 + v[1]**2)

def vector_projection(u, v):
    """Proyecci√≥n del vector u sobre v"""
    v_norm_squared = v[0]**2 + v[1]**2
    scalar_projection = np.dot(u, v) / v_norm_squared
    return scalar_projection * v

# Visualizaci√≥n gr√°fica de la suma de vectores
def plot_vector_addition(u, v):
    """Representa gr√°ficamente la suma de vectores usando el m√©todo cabeza-cola"""
    result = vector_addition(u, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Dibujar los vectores
    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='v')
    plt.quiver(0, 0, result[0], result[1], angles='xy', scale_units='xy', scale=1, color='red', label='u+v')

    # Configurar l√≠mites y etiquetas
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('Suma de vectores (M√©todo cabeza-cola)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Visualizaci√≥n gr√°fica de la multiplicaci√≥n por escalar
def plot_scalar_multiplication(alpha, v):
    """Representa gr√°ficamente la multiplicaci√≥n por escalar"""
    scaled_v = scalar_multiplication(alpha, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Dibujar el vector original y el escalado
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='v')
    plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='red', label=f'{alpha}v')

    # Configurar l√≠mites y etiquetas
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('Multiplicaci√≥n por escalar')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Visualizaci√≥n gr√°fica de la proyecci√≥n
def plot_vector_projection(u, v):
    """Representa gr√°ficamente la proyecci√≥n de un vector sobre otro"""
    proj = vector_projection(u, v)

    plt.figure(figsize=(8, 8))
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)

    # Dibujar los vectores
    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='blue', label='u')
    plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='green', label='v')
    plt.quiver(0, 0, proj[0], proj[1], angles='xy', scale_units='xy', scale=1, color='red', label='Proyecci√≥n de u sobre v')

    # Configurar l√≠mites y etiquetas
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.legend()
    plt.title('Proyecci√≥n de un vector sobre otro')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Ejemplo de uso
if __name__ == "__main__":
    # Definir vectores
    u = np.array([3, 2])
    v = np.array([1, 4])

    print("Vector u:", u)
    print("Vector v:", v)

    # Suma de vectores
    u_plus_v = vector_addition(u, v)
    print("Suma de vectores u + v:", u_plus_v)

    # Multiplicaci√≥n por escalar
    alpha = 2
    scaled_u = scalar_multiplication(alpha, u)
    print(f"Vector u multiplicado por {alpha}:", scaled_u)

    # Norma de un vector
    norm_u = vector_norm(u)
    print("Norma del vector u:", norm_u)

    # Proyecci√≥n de u sobre v
    proj_u_on_v = vector_projection(u, v)
    print("Proyecci√≥n de u sobre v:", proj_u_on_v)

    # Visualizaci√≥n gr√°fica
    plot_vector_addition(u, v)
    plot_scalar_multiplication(alpha, u)
    plot_vector_projection(u, v)
```
